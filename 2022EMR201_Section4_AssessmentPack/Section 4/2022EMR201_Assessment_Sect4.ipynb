### Question 4: Simulating a Mass-Spring-Pendulum System (11 marks) <a id="A_2_q_4"></a>

**Scenario:** A free body diagram of a Mass-Spring-Pendulum system is shown in the figure below. A block of mass $m$ is fastened to a massless rod of length $L$. The massless rod is connected to an angular spring. When the block is moved from its resting position (i.e. $\theta$=0 rad) and let go from position $\theta=2\pi\ rad$, it rotates back and forth following an angular path. The primary forces acting on the block are defined as $F_s$ the spring force and $F_d$ the damping force resulting from friction.  

<img src="Torque_spring_sys.png" width="700" height="500">

You have been tasked with modelling and solving this system mathematically in order to describe how the angle $\theta$ varies as a function of time. The following table provides a summary of the variables.

| Variable | Description | 
| --- | --- |
|  $m$ | Block Mass |
|  $F_s$ | Spring Force |
|  $F_d$ | Damping Force |
|  $b$ | Damping Coefficient |
|  $k$ | Spring Constant|
|  $L$ | Rod Length|
|  $\theta(t)$ | Instantaneous Position |
|  $\dot{\theta}(t)$ | Instantaneous Velocity |


In order to mathematically model the Mass-Spring-Pendulum System, you will need to familiarize yourself with Newton's Law of Motions. 

Newton's Second Law of Motion states that the vector sum of all forces acting on an object is equal to the mass of the object multiplied by its acceleration. This is defined mathematically as:

\begin{equation*}
\sum F = ma
\label{eq:newton_2} \tag{1}
\end{equation*}


Since the normal force equals the gravitational force acting on the block, one only needs to summarize the forces acting on the block along its path of movement. Eq. $\eqref{eq:sum_forces}$ demonstrates the summation of all the forces in the x-direction for a translational movement. 

\begin{equation*}
\sum F = F - kx - b\ddot x
\label{eq:sum_forces} \tag{2}
\end{equation*}


However, the block in our Mass-Spring-Pendulum system is moving in an angular motion. Therefore Newton's Second Law given in eq. $\eqref{eq:newton_2}$ should be modified for rotational movement as shown in eq. $\eqref{eq:moments}$.

\begin{equation*}
\sum M = I\alpha
\label{eq:moments} \tag{3}
\end{equation*}


Where $M$ is the sum of moments, $I$ is the moment of inertia and $\alpha$ is the angular acceleration.

The expression in eq. $\eqref{eq:sum_forces}$ can also be translated into an expression for rotational movement by using the angular displacement of the block denoted by $\theta$, instead of the linear displacement, $x$. Also, the initial displacement of the block from equilibrium will cause the mass to move instead of a constant applied force. Therefore, the only forces acting on the mass along the path of movement are the damping force and the spring force. This new expression is shown in eq. $\eqref{eq:sum_moments}$.   

\begin{equation*}
\sum M = - k\theta - b\dot\theta
\label{eq:sum_moments} \tag{4}
\end{equation*}

(Note that although these forces are proportional to the movement of the block, they act against the direction of motion.)


Furthermore, $I$ the moment of inertia and $\alpha$ the acceleration of the ball can also be defined mathematically as,

\begin{equation*}
I = mL^2
\label{eq:inertia} \tag{5}
\end{equation*}

\begin{equation*}
\alpha = \ddot\theta
\label{eq:alpha} \tag{6}
\end{equation*}

respectively. Take note that acceleration has also been defined with respect to the angle $\theta$.

Combining equations $\eqref{eq:moments}$ - $\eqref{eq:alpha}$, renders the following second order differential equation for the angular accelaration:


\begin{equation*}
\ddot\theta = \frac{-(b\dot\theta +k\theta)}{(mL^2)}
\label{eq:angular_accelaration} \tag{7}
\end{equation*}

In order to solve this second-order differential equation, one must simplify eq. $\eqref{eq:angular_accelaration}$ to a first order differential equation. This is achieved by using the substitution method. Setting
$\theta$ equal to $\theta_1$ and $\dot\theta$ equal to $\theta_2$ in eq. $\eqref{eq:angular_accelaration}$ renders the ordinary differential equation shown below. Eq. $\eqref{eq:final_angular_accelaration}$ can then be solved using discreet methods such as Euler. 

\begin{equation*}
\dot\theta_2 = \frac{-(b\theta_2 +k\theta_1)}{(mL^2)}
\label{eq:final_angular_accelaration} \tag{8}
\end{equation*}
